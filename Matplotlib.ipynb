```python
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3],[1,8,27])
```

```python
plt.plot([1,2,3],[1,8,27])
plt.title("myplot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
x=[1,2,3,4]
y=[2,4,6,8]
y2=[3,6,9,12]
plt.title("myplot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y,label="2x",linewidth=3,color="b")
plt.plot(x,y2,label="3x")
plt.legend()
plt.show()
```

```python
x=[1,2,3,4]
y=[2,4,6,8]
y2=[3,6,9,12]
plt.title("myplot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.scatter(x,y,label="2x",linewidth=3,color="r")
plt.scatter(x,y2,label="3x")
plt.legend()
plt.show()
```

```python
x=[1,2,3,4]
y=[2,4,6,8]
y2=[3,6,9,12]
plt.title("myplot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y,label="2x",linewidth=3,color="b")
plt.hist(y2,label="3x")
plt.legend()
plt.show()
```
